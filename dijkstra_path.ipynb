```python
import pandas as pd
np.set_printoptions(threshold=sys.maxsize)


GrafoV = pd.read_csv('GrafoV.csv')
pd.set_option('display.max_rows', 1000)
display(GrafoV)
```

```python
#Libreria para los grafos
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph() # Se crea un grafo sin valores
for row in GrafoV.iterrows(): #iterrows() recorre en iteraci√≥n las filas de dataframe como pares
    G.add_edge(row[1]["origen"],row[1]["destino"],distancia=row[1]["distancia(m)"]) #se llena el grafo con los valores de la
                                                                                    #tabla de datos
```

```python
G.nodes(data=True)
```

```python
nx.draw_kamada_kawai(G,ax=plt.axes((4,4+.25*2,3,3)),font_size=18,width=2,with_labels=True, node_size=3000)
```

```python
from time import time

def mostrar_camino(camino): #nos indica el punto de origen y el punto destino y la distancia que hay entre ellos
    distancia_total = 0 
    tiempo_inicial = time()
    
    for i in range(len(camino)-1):#len() devuelve el numero de elementos de una lista 
        origen = camino[i] 
        destino = camino[i+1]
        distancia = G[origen][destino]["distancia"] # analizando la distancia del grafo entre el origen y el destino
    
    
        distancia_total = distancia_total+distancia #Calculando la distancia total recorrida
        print("Origen: ", origen+" -  Destino: ",destino +" - Distancia: ",distancia)
        
    tiempo_final = time()
    tiempo_total = tiempo_final - tiempo_inicial # Se calcula el tiempo de ejecuci√≥n de la funci√≥n
        
    print("\nDistancia total en metros[m]: ",distancia_total)
    print("\nTiempo de Ejecuci√≥n en segundos[s]",tiempo_total)
```

```python
def grafo_camino(camino): #Esta funci√≥n nos muestra la ruta m√°s corta a traves de un grafo
    print(camino)
    pos = nx.kamada_kawai_layout(G) #mantiene una forma definida del grafo y hace que no se genere otro automaticamente
    
    #Grafo ya creado previamente
    nx.draw(G,pos=pos,ax=plt.axes((4,4+.25*2,3,3)),node_color='purple', edge_color='gray',
            font_size=24, width=1, with_labels=True, node_size=2000, alpha=0.5)
    
    #creamos un nuevo grafo el cual tendra el camino m√°s corto
    GC = nx.DiGraph()
    for i in range(len(camino)-1): # len() devuelve el numero de elementos de una lista 
        GC.add_edge(camino[i],camino[i+1])
    
    nx.draw(GC,pos=pos,ax=plt.axes((4,4+.24*2,3,3)),node_color='purple', edge_color='blue',
            font_size=24, width=3, with_labels=True, node_size=2000,)
    
    plt.show()
```
